```python
import pandas as pd
from Bio import SeqIO
import os
```

```python
direct = '/home/emulciber/MitoClub/bacteria/Body/1Raw/all_genes_aa' #ввести название директории с файлами аминокислот генов
dirname = os.listdir(direct)
```

```python
bacteria_test = pd.DataFrame()
```

```python
species = []
gene = []
aaseq = []
oid = []


for file in dirname:
    try:
        for seq_record in SeqIO.parse((os.path.join(direct, file)), 'fasta'):
            # oid.append(file.split(' ')[0])
            gene.append(file.split('.')[0]) # сюда воткнуть название файла
            aaseq.append(str(seq_record.seq)) # получение последовательности из фасты
            # получение вида из дескрипшна
            desc = seq_record.description
            oid.append(desc.split(' ')[0])
            if str(desc.split('[')[-1]).startswith('Ga') == True:
                species.append(' '.join(desc.split('[')[-1].split(' : ')[-1].split(' ')[0:2]))
            else:
                species.append(' '.join(desc.split('[')[-1].split(' ')[0:2]))
    except:
        pass
    

bacteria_test['gene_oid'] = oid
bacteria_test['Species'] = species
bacteria_test['Gene'] = gene
bacteria_test['Sequence'] = aaseq
bacteria_test
```

```python
bacteria_test.to_csv('/home/emulciber/MitoClub/bacteria/Body/2Derived/bacteria_test.csv', sep='\t')
```

```python
## ЛАЖА

species_set = []
for i in range(len(species_list)):
    for k in range(i+1, len(species_list)):
        if species_list[i].split(' ')[0] == species_list[k].split(' ')[0] and oxygen_list[i] != oxygen_list[k]:
            species_set.append(species_list[i])
            species_set.append(species_list[k])
            
species_set = set(species_set)
```

```python
from ete3 import NCBITaxa 
ncbi = NCBITaxa()
```

```python
species_list = bacteria['Species']
taxonomy_list = []

for species in species_list:
    name2taxid = list(ncbi.get_name_translator([species]).values())
    if len(name2taxid) == 0:
        taxonomy_list.append(name2taxid)
        continue
    lineage = ncbi.get_lineage(name2taxid[0][0])
    names = ncbi.get_taxid_translator(lineage)
    taxonomy_list.append([names[taxid] for taxid in lineage])
```

```python
bacteria['Taxonomy'] = taxonomy_list
```

```python
bacteria['Phylum'] = bacteria['Taxonomy'].apply(lambda x: x[4] if len(x) != 0 else None)
```
